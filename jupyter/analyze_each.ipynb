```python
df1 = pd.read_csv(target1)
df2 = pd.read_csv(target2)
df1['score2'] = df2.score
df1
```

```python
losecase = df1[df1.score * 0.8 > df1.score2]
losecase
```

```python
wincase = df1[df1.score < df1.score2 * 0.8]
wincase
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
f, ax = plt.subplots(1, 2, figsize=(18, 8))
sns.distplot(losecase.maxTime, bins=10, ax=ax[0])
sns.distplot(wincase.maxTime, bins=10, ax=ax[1])
plt.show()
```

```python
label = ['coins', 'noteTime', 'numMachines']
f, ax = plt.subplots(len(label), 2, figsize=(18, 18))

for i, name in enumerate(label):
    sns.distplot(losecase[name], bins=8, ax=ax[i, 0])
    sns.distplot(wincase[name], bins=8, ax=ax[i, 1])
plt.show()
```

```python
f, ax = plt.subplots(1, 2, figsize=(18, 8))
for i, case in enumerate([losecase, wincase]):
    casedata = case.maxTime * 0.2 / case.noteTime / case.numMachines
    ax[i].set_xlim(0, 50)
    sns.distplot(casedata, bins=50, ax=ax[i])
plt.show()
```
