```python
import os
import pandas as pd

base_dir = '../results/'

begin = 1001
end = 2000
```

```python
df = []
names = []

exclude = ['result_0318_144659.csv']

# seedが[begin,end]の範囲内のケースを出力したもののみ比較
for name in os.listdir(base_dir):
#     if name in exclude:
#         continue
    path = base_dir + name
    csv = pd.read_csv(path, index_col='seed')
    index = csv.index
    if index[0] != begin or index[-1] != end:
        continue
    names.append(name)
    df.append(csv.score)

df = pd.concat(df, axis=1)
df.columns = names
df
```
