```python
import pandas as pd

filepath = '../stat.csv'
df = pd.read_csv(filepath)
df
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
(df.T / df.trueExp - 1).T.abs().mean()[1:].plot(figsize=(15, 5), grid=True)
```

```python
(df.T / df.trueExp - 1).T.std()[1:].plot(figsize=(15, 5), grid=True)
```

```python
import seaborn as sns
diff = (df.T / df.trueExp - 1).T

fig, ax = plt.subplots(1, 3, figsize=(18, 5))
sns.distplot(diff.exp5.fillna(0), ax=ax[0])
sns.distplot(diff.exp15.fillna(0), ax=ax[1])
sns.distplot(diff.exp30.fillna(0), ax=ax[2])

plt.show()
```

```python
# 過小評価されがち
(diff.exp5 < 0).value_counts()
```
